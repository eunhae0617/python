```python
import requests
from urllib.parse import urlparse
import pandas as pd
import time

title_text = []
link_text = []
description_text = []
blogger_name = []

def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/blog?query=" + keyword \
          + "&display=" + str(display) \
          + "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
           headers={"X-Naver-Client-Id":"7Ze0Rx9lT6OsC49kTPOV",
                   "X-Naver-Client-Secret":"u9YViJjkkc"})
    return result.json()


def call_and_print(keyword, page):
    json_obj = get_api_result(keyword, 100, page)
    for item in json_obj['items']:
        time.sleep(0.5)
        title = item['title'].replace("<b>","").replace("</b>","")
        description = item['description'].replace("<b>","").replace("</b>","")
        
        title_text.append(title)
        link_text.append(item['link'])
        description_text.append(description)
        blogger_name.append(item['bloggername'])
        
             
call_and_print("코로나", 3)

results = {'title':title_text,'link':link_text,'description':description_text,'bloggername':blogger_name}
data = pd.DataFrame(results)
data.to_csv('d:/corona.csv')
data.head(10)
```
